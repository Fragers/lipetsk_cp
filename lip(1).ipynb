```python
import numpy as np
import torch
import open_clip
from open_clip import tokenizer
import os
from PIL import Image
import pandas as pd
from tqdm import tqdm
```

```python
df_tr = pd.read_csv("/content/drive/MyDrive/university/translated.csv")
```

```python
df_tr.head()
```

```python
model, _, preprocess = open_clip.create_model_and_transforms(
    'ViT-H-14', 
    pretrained='laion2b_s32b_b79k', 
    device=torch.device("cuda:0"))
model.eval();
```

```python
images = []
img_dir = "/content/drive/MyDrive/university/test"
img_ids = []
vectors = []
for img_name in tqdm(os.listdir(img_dir)):
    image = Image.open(os.path.join(img_dir, img_name)).convert("RGB")
    img_ids.append(img_name.split('.')[0])
    pre = preprocess(image)
    pre = pre.to("cuda:0")
    # print(pre.shape)
    with torch.no_grad():
        image_features = model.encode_image(pre[None, ...]).float()
        vectors.append(image_features)
```

```python
df_img = pd.DataFrame()
```

```python
list_img_vecs = list(map(lambda x: x.cpu().reshape(-1).numpy().tolist(), vectors))
```

```python
df_img["id"] = img_ids
df_img["vec"] = list_img_vecs
df_img
```

```python
df_img.to_csv("/content/drive/MyDrive/university/img_df.csv", index=False)
```

```python
text_vectors = []
for en in tqdm(df_tr["description_en"]):
    tok = tokenizer.tokenize("This is " + en)
    tok = tok.to("cuda:0")
    with torch.no_grad():
        # print(tok)
        text_features = model.encode_text(tok).float()
        text_vectors.append(text_features)
```

```python
# text_vectors
df_txt = pd.DataFrame()
list_txt_vecs = list(map(lambda x: x.cpu().reshape(-1).numpy().tolist(), text_vectors))
df_txt["id"] = df_tr["id"]
df_txt["vec"] = list_txt_vecs
df_txt
```

```python
df_txt.to_csv("/content/drive/MyDrive/university/df_txt.csv", index=False)
```

```python
import numpy as np
import torch
# import open_clip
# from open_clip import tokenizer
import os
from PIL import Image
import pandas as pd
from tqdm import tqdm
```

```python
df_txt = pd.read_csv("/content/drive/MyDrive/university/df_txt.csv")
df_img = pd.read_csv("/content/drive/MyDrive/university/img_df.csv")
```

```python
img_ids = df_img["id"].values
img_vecs = df_img["vec"].values

txt_ids = df_txt["id"].values
txt_vecs = df_txt["vec"].values 
```

```python
def parse_df_vecs(vec):
    return np.array(list(map(float, vec[1:-1].split(","))), dtype=float)[None, ...]
```

```python
img_vecs_2 = list(map(parse_df_vecs, img_vecs))
img_vecs_2 = np.concatenate(img_vecs_2, axis=0)
# img_vecs_2.shape

txt_vecs_2 = list(map(parse_df_vecs, txt_vecs))
txt_vecs_2 = np.concatenate(txt_vecs_2, axis=0)
txt_vecs_2.shape
```

```python
# img_vecs_2 = np.concatenate([x.cpu().numpy() for x in vectors], axis=0)
# txt_vecs_2 = np.concatenate([x.cpu().numpy() for x in text_vectors], axis=0)
```

```python
# image_features /= img_vecs_2.norm(dim=-1, keepdim=True)
# text_features /= txt_vecs_2.norm(dim=-1, keepdim=True)
image_featurs = img_vecs_2 / np.linalg.norm(img_vecs_2, axis=1)[..., None]
text_featurs = txt_vecs_2 / np.linalg.norm(txt_vecs_2, axis=1)[..., None]
# print(text_featurs.shape)
similarity = image_featurs @ text_featurs.T
print(similarity.shape)
```

```python
from scipy.optimize import linear_sum_assignment
```

```python
# similarity_np = similarity.cpu().numpy()
row_ind, col_ind = linear_sum_assignment(similarity)
```

```python
pairs = list(zip(row_ind, col_ind))
# pairs
```

```python
similarity
```

```python
pairs = []
for i in range(900):
    mx = np.argmax(similarity)
    r = mx // 900
    c = mx % 900
    pairs.append((r, c))
    similarity[:, c] = -1
    similarity[r, :] = -1
```

```python
img_ids_fin = list(map(lambda x: x[0], pairs))
txt_ids_fin = list(map(lambda x: x[1], pairs))
```

```python
txt_start_id = df_tr['id'].tolist()
# txt_start_id
```

```python
mapped_ids_txt_fin = [txt_ids[i] for i in txt_ids_fin]
mapped_ids_img_fin = [img_ids[i] for i in img_ids_fin]
```

```python
fin = pd.DataFrame()
fin["id"] = mapped_ids_txt_fin
fin["object_img"] = mapped_ids_img_fin
fin.to_csv("ans_fixed_ids.csv", index=False)
```
